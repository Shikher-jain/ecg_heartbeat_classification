```python
import numpy as np
X = np.load("../data/raw/X_raw.npy")
y = np.load("../data/raw/y_raw.npy")
```

```python
from src.preprocess import preprocess_dataset
X_proc, y_proc = preprocess_dataset(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(
    X_proc.reshape(-1,187,1), y_proc, test_size=0.2, stratify=y_proc
)
```

```python
from src.model import build_cnn_model
model = build_cnn_model(input_length=187)
```

```python
history = model.fit(
    X_train, y_train, epochs=30, batch_size=32,
    validation_data=(X_val, y_val)
)
```

```python
model.save("../saved_models/ecg_cnn.h5")
```

```python
preds = model.predict(X_val).reshape(-1)
from sklearn.metrics import accuracy_score, precision_score, recall_score
preds_label = (preds>0.5).astype(int)
print("Accuracy:", accuracy_score(y_val, preds_label))
print("Precision:", precision_score(y_val, preds_label))
print("Recall:", recall_score(y_val, preds_label))
```
